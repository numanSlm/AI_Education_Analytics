```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("final.csv")
```

```python
cdf = df[df["age"] == "c"]
```

```python
ndf = pd.concat([df,cdf,cdf,cdf])
```

```python
sns.countplot(ndf,x='emotion',hue='age')
plt.savefig("child.svg")
```
